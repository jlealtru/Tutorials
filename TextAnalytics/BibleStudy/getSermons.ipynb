### Secure Sermon Text

For the purposes of this text analytics series, we are going to use sermons as our core dataset. We can build our sermon bot using this data, we can build a network of sermons to bible passages, and we can measure the similarity of messages within a sermon to the actual passages of the bible. 

Our initial sermon data will come from:

- [preaching.com](https://www.preaching.com/sermons/)
- [Desiring God](https://www.desiringgod.org)
- [Sermon Index](http://www.sermonindex.net/modules/articles/)

```python
preach_url = 'https://www.preaching.com/sermons/'
preach_page_url = 'https://www.preaching.com/sermons/page/{}/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

soup = BeautifulSoup(requests.get(preach_url, headers=headers).content, 'html.parser')
```

```python
def parse_articles(article):
    title = getattr(article.find('h2', {'class': 'cb-post-title'}), 'text', None)
    article_url = article.find('a').get('href', 'NO_URL')
    return (title, article_url)
```

```python
master_list = []

def get_next_page(soup):
    pg_numbers = soup.find('ul', {'class': 'page-numbers'})
    next_pg_url = pg_numbers.find('span', {'class': 'page-numbers current'}).findNext('li').find('a').get('href')
    if 'page' in next_pg_url:
        return next_pg_url
    else:
        return None
    
def iter_articles(url, cur_page=1):
    print(url)
    global master_list
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    soup = BeautifulSoup(requests.get(url, headers=headers).content, 'html.parser')
    
    for article in soup.find('div', {'id': 'main'}).findAll('article'):
        title, article_url = parse_articles(article)
        print(title)
    # next_pg_url = get_next_page(soup)

        tmpdf = pd.DataFrame({'title': title,
                             'article_url': article_url}, index=[0])

        master_list.append(tmpdf)

    if cur_page <= 157:
        cur_page = cur_page + 1
        print(cur_page)
        next_url = preach_page_url.format(cur_page)
        iter_articles(next_url, cur_page=cur_page)
```

```python
all_sermons = pd.concat(master_list)
```

### Get the actual sermon text content

Iterate over each sermon link and pull the body of text

```python
def create_soup(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    soup = BeautifulSoup(requests.get(url, headers=headers).content, 'html.parser')
    return soup

def get_author(soup):
    # subset author tags, get author text and author url
    author = soup.find('div', {'class': 'cb-author cb-byline-element'})
    author_text = getattr(author, 'text')
    author_url = author.find('a')['href']
    return (author_text, author_url)

def clean_sermon_text(txt):
    while '  ' in txt:
        txt = txt.replace('  ', ' ')
    # filter ords
    txt = ''.join([char for char in txt if ord(char) <= 128])
    return txt

def get_sermon_text(soup):
    sermon_text = soup.find('section', {'class': 'cb-entry-content entry-content clearfix'}).text
    sermon_text = clean_sermon_text(sermon_text)
    return sermon_text
```

```python
superdf.head(1)
seen_urls = superdf['url'].unique().tolist()
```

```python
import sys
bad_rows = []

for index, row in all_sermons.loc[~all_sermons['article_url'].isin(seen_urls)].reset_index().iterrows():
    sys.stdout.write('\rINDEX: {}'.format(index))
    sys.stdout.flush()
    url = row['article_url']
    # print('INDEX: {}\nURL: {}'.format(index, url))
    title = row['title']
    try:
        soup = create_soup(url)
        author_text, author_url = get_author(soup)
        sermon_text = get_sermon_text(soup)
        tmprow = pd.DataFrame({'url': url,
                              'title': title,
                              'author_text': author_text,
                              'author_url': author_url,
                              'sermon_text': sermon_text}, index=[0])
        masterdf.append(tmprow)
    except (AttributeError, TypeError) as e:
        bad_rows.append(url)
```
